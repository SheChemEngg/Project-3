```python
# ['job_state', 'job_region', 'Sector', 'Company_Age', 'Avg_Company_Size', 'Job_Title', 'Avg_Salary', 
# 'Max_Salary', 'Usr_Rating', 'Python_y_n', 'R_y_n', 'Spark_y_n', 'AWS_y_n', 'Excel_y_n']

lang_list = ['Python_y_n', 'R_y_n', 'Spark_y_n', 'AWS_y_n', 'Excel_y_n']
header_list = ['Python', 'R', 'Spark', 'AWS', 'Excel']

def language(lang, title, ax):

    count1 = glassdoor[lang].value_counts()[0]
    count2 = glassdoor[lang].value_counts()[1]
    if count1 > count2:
          color_palette = {1: 'royalblue', 0: 'darkorange'}
    else:
          color_palette = {1: 'darkorange', 0: 'royalblue'}

    mask = glassdoor['Usr_Rating'] > 0
    if mask.any():
        plot = sns.scatterplot(x = 'Usr_Rating', y='Avg_Salary', data = glassdoor[mask], hue = lang, palette = color_palette, ax = ax)
    
    if plot.get_legend():
        legend_labels = plot.get_legend().get_texts()
        for label in legend_labels:
            label.set_text('No' if label.get_text() == '0' else 'Yes')
        plot.get_legend().set_title('')
    
    gldr_lang = glassdoor[glassdoor[lang].isin([1])]
    gldr_lang_sal = int(gldr_lang.Avg_Salary.median())
    # print(f'Median Salary with knowledge of {title}: ${gldr_lang_sal}k')
    plot.set_yticks([gldr_lang_sal])
    plot.set_xticks([3.7])
    plot.set_title(title)

fig, axs = plt.subplots(nrows = 2, ncols = 3, figsize = (15, 10))
fig.suptitle('Annual Salary vs. Company Rating for jobs requiring Programming Language/Application')
fig.subplots_adjust(hspace = 0.5, wspace = 0.5)

i = j = counter = 0
for l in range(len(lang_list)):
    language(lang_list[l], header_list[l], axs[i][j])
    j += 1
    counter += 1
    if counter > 4:
        plt.delaxes(axs[i][j])
        break

    if j > 2:
        i += 1; j = 0
plt.savefig("C:\A_Sheena\Cyber\Module_16Pj3\Project_3\Output_data\Fig11.png")
```

```python
# ['job_state', 'job_region', 'Sector', 'Company_Age', 'Avg_Company_Size', 'Job_Title', 'Avg_Salary', 
# 'Max_Salary', 'Usr_Rating', 'Python_y_n', 'R_y_n', 'Spark_y_n', 'AWS_y_n', 'Excel_y_n']

unique_states = glassdoor['job_state'].unique()
state_job_types = {}

for state in unique_states:
    # Filter the data for the current state
    state_data = glassdoor[glassdoor['job_state'] == state]

    # Count the number of each Job_Title for the current state and sort in descending order
    job_counts = state_data['Job_Title'].value_counts().sort_values(ascending=False).to_dict()

    # Store the job counts in the state_job_types dictionary
    state_job_types[state] = job_counts

# Order the dictionary by ascending keys
sorted_state_job_types = dict(sorted(state_job_types.items()))
state_job_types = sorted_state_job_types

print(state_job_types.keys()) 
print()

# Print the state_job_types dictionary
print(state_job_types)
```

```python
# # ['job_state', 'job_region', 'Sector', 'Company_Age', 'Avg_Company_Size', 'Job_Title', 'Avg_Salary', 
# # 'Max_Salary', 'Usr_Rating', 'Python_y_n', 'R_y_n', 'Spark_y_n', 'AWS_y_n', 'Excel_y_n']
# *****
# Transform the data structure
transformed_data = []

for state_code, job_data in state_job_types.items():
    # Create a dictionary for the current state
    state_info = {
        "state_code": state_code,
        "job_titles": [],
        "job_numbers": []
    }

    # Sort job titles by the number of jobs in descending order
    sorted_jobs = sorted(job_data.items(), key=lambda x: x[1], reverse=True)

    for job_title, job_number in sorted_jobs:
        state_info["job_titles"].append(job_title)
        state_info["job_numbers"].append(job_number)

    transformed_data.append(state_info)

# print(transformed_data)
state_job_types = transformed_data
print(state_job_types)
```

```python
# ['job_state', 'job_region', 'Sector', 'Company_Age', 'Avg_Company_Size', 'Job_Title', 'Avg_Salary', 
# 'Max_Salary', 'Usr_Rating', 'Python_y_n', 'R_y_n', 'Spark_y_n', 'AWS_y_n', 'Excel_y_n']

unique_states = glassdoor['job_state'].unique()
state_sector_types = {}

for state in unique_states:
    # Filter the data for the current state
    state_data = glassdoor[glassdoor['job_state'] == state]

    # Calculate the number of jobs in a sector and average Avg_Salary for the current state
    sector_counts = state_data['Sector'].value_counts()
    sector_avg_salaries = round(state_data.groupby('Sector')['Avg_Salary'].mean(), 1)

    # Combine the counts and average salaries into a list
    job_counts = [
        {
            'Sector': sector,
            'Number_of_Jobs': sector_counts.get(sector, 0),  # Get the count or 0 if not found
            'Average_Avg_Salary': sector_avg_salaries.get(sector, 0.0),  # Get the average or 0.0 if not found
        }
        for sector in sector_counts.index
    ]

    # Store the job counts in the state_job_types dictionary
    state_sector_types[state] = job_counts

sorted_state_sector_types = dict(sorted(state_sector_types.items()))
state_sector_types = sorted_state_sector_types    
    
# Print the state_job_types dictionary
print(state_sector_types)
```

```python
# state_sector_types values into list form

result = {}

for state, data in state_sector_types.items():
    sectors = []
    average_salaries = []
    num_jobs = []

    for sector_data in data:
        sectors.append(sector_data['Sector'])
        average_salaries.append(sector_data['Average_Avg_Salary'])
        num_jobs.append(sector_data['Number_of_Jobs'])

    # Sort the lists by the number of jobs in descending order
    sorted_data = sorted(zip(sectors, average_salaries, num_jobs), key=lambda x: x[2], reverse=True)
    sectors, average_salaries, num_jobs = zip(*sorted_data)

    result[state] = {
        'Sectors': list(sectors),
        'Average_Avg_Salary': list(average_salaries),
        'Number_of_Jobs': list(num_jobs)
    }

state_sector_types = result
print(state_sector_types)
```
