```python
import math

def size_average(company_size):
    
    if company_size == 'Unknown' or company_size == '-1':
        return int('0')
    
    else:
        # Split the string by space
        parts = company_size.split()
        
        if '+' in parts[0]:
            return int(parts[0].replace('+', ''))
        else:
            # Average of the two numbers
            average = int((int(parts[0]) + int(parts[2])) / 2)
            return average

# Apply the function to create the 'Avg_Company_Size' column
glassdoor['Avg_Company_Size'] = glassdoor['Company_Size'].apply(size_average)

glassdoor.tail()
```

```python
glassdoor.drop('Company_Size', axis=1, inplace=True)
glassdoor.info()
```

```python

def simplify_job_title(title):
    title = title.lower()
    if 'data' in title and ('scientist' in title or 'science' in title or 'analytics' in title):
        return 'Data Scientist'
    elif 'data' in title and 'engineer' in title:
        return 'Data Engineer'
    elif 'machine' in title and 'learning' in title:
        return 'Machine Learning Engineer'
    elif 'analyst' in title:
        return 'Data Analyst'
    elif 'director' in title:
        return 'Director'
    elif 'manager' in title:
        return 'Manager'
    else:
        return 'Research Scientist'

# Apply the function to the 'Job_Title' column
glassdoor['Job_Title'] = glassdoor['Job_Title'].apply(simplify_job_title)

glassdoor['Job_Title'].unique()
```

```python
# Change glassdoor['Competitors_y_n'] = '-1' to '0'
glassdoor['Competitors_y_n'] = glassdoor['Competitors_y_n'].replace('-1', '0')

# Change others to '1'
glassdoor.loc[glassdoor['Competitors_y_n']!='0', 'Competitors_y_n'] = '1' 

glassdoor['Competitors_y_n'] = glassdoor['Competitors_y_n'].astype(int)
glassdoor['Competitors_y_n'].head()
```

```python
# 5 Regions according to National Geographic

regions = {
    'West': "WA-MT-OR-ID-WY-CA-NV-UT-CO-AK-HI",
    'Southwest': "AZ-NM-TX-OK",
    'Midwest': "ND-SD-NE-KS-MN-IA-MO-WI-IL-MI-IN-OH",
    'Southeast': "KY-WV-VA-AR-TN-NC-SC-GA-AL-MS-LA-FL",
    'Northeast': "PA-NY-VT-ME-NH-MA-RI-CT-NJ-DE-MD-DC"
    }

# Function to map state to region
def state_to_region(state):
    for region, states in regions.items():
        if state in states.split('-'):
            return region
    # return None  # Return None if no region is found

# Create the 'job_region' column by applying the mapping function
glassdoor['job_region'] = glassdoor['job_state'].apply(state_to_region)

glassdoor['job_region'].head()
```

```python
# ['job_state', 'job_region', 'Sector', 'Company_Age', 'Avg_Company_Size', 'Job_Title', 'Avg_Salary', 
# 'Max_Salary', 'Usr_Rating', 'Python_y_n', 'R_y_n', 'Spark_y_n', 'AWS_y_n', 'Excel_y_n']

unique_states = glassdoor['job_state'].unique()
state_job_types = {}

for state in unique_states:
    # Filter the data for the current state
    state_data = glassdoor[glassdoor['job_state'] == state]

    # Count the number of each Job_Title for the current state and sort in descending order
    job_counts = state_data['Job_Title'].value_counts().sort_values(ascending=False).to_dict()

    # Store the job counts in the state_job_types dictionary
    state_job_types[state] = job_counts

# Order the dictionary by ascending keys
sorted_state_job_types = dict(sorted(state_job_types.items()))
state_job_types = sorted_state_job_types

# ---------------------------------------------------------------------

# Transform the data structure
transformed_data = []

for state_code, job_data in state_job_types.items():
    # dictionary for current state
    state_info = {
        "state_code": state_code,
        "job_titles": [],
        "job_numbers": []
    }

    # Sort job titles by the number of jobs in descending order
    sorted_jobs = sorted(job_data.items(), key=lambda x: x[1], reverse=True)

    for job_title, job_number in sorted_jobs:
        state_info["job_titles"].append(job_title)
        state_info["job_numbers"].append(job_number)

    transformed_data.append(state_info)

# Rename transformed_data
state_job_types = transformed_data
print(state_job_types)
```

```python
# ['job_state', 'job_region', 'Sector', 'Company_Age', 'Avg_Company_Size', 'Job_Title', 'Avg_Salary', 
# 'Max_Salary', 'Usr_Rating', 'Python_y_n', 'R_y_n', 'Spark_y_n', 'AWS_y_n', 'Excel_y_n']

unique_states = glassdoor['job_state'].unique()
state_sector_types = {}

for state in unique_states:
    # Filter the data for the current state
    state_data = glassdoor[glassdoor['job_state'] == state]

    # Calculate the number of jobs in a sector and average Avg_Salary for the current state
    sector_counts = state_data['Sector'].value_counts()
    sector_avg_salaries = round(state_data.groupby('Sector')['Avg_Salary'].mean(), 1)

    # Combine the counts and average salaries into a list
    job_counts = [
        {
            'Sector': sector,
            'Number_of_Jobs': sector_counts.get(sector, 0),  # Get the count or 0 if not found
            'Average_Avg_Salary': sector_avg_salaries.get(sector, 0.0),  # Get the average or 0.0 if not found
        }
        for sector in sector_counts.index
    ]

    # Store the job counts in the state_job_types dictionary
    state_sector_types[state] = job_counts

sorted_state_sector_types = dict(sorted(state_sector_types.items()))
state_sector_types = sorted_state_sector_types    
    
# ----------------------------------------------------------------

# Transform state_sector_types values into list form
result = {}

for state, data in state_sector_types.items():
    sectors = []
    average_salaries = []
    num_jobs = []

    for sector_data in data:
        sectors.append(sector_data['Sector'])
        average_salaries.append(sector_data['Average_Avg_Salary'])
        num_jobs.append(sector_data['Number_of_Jobs'])

    # Sort in descending order by job counts
    sorted_data = sorted(zip(sectors, average_salaries, num_jobs), key=lambda x: x[2], reverse=True)
    sectors, average_salaries, num_jobs = zip(*sorted_data)

    result[state] = {
        'Sectors': list(sectors),
        'Average_Avg_Salary': list(average_salaries),
        'Number_of_Jobs': list(num_jobs)
    }

state_sector_types = result
print(state_sector_types)
```
